3.4.1 Logistic Regression

Here, we will do a logistic regression problem, starting with the graphing of a sigmoid funciton

```python
#imports
import numpy as np
import matplotlib.pyplot as plt

#define function

def sigmoid(x):
  return 1 / (1 + np.exp(-x))

#plot the graph
x = np.linspace(-10, 10, 100)
y = sigmoid(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.title("Sigmoid Function")
plt.grid(True)
plt.show()
```

Now compute the gradient descent

```python
# Compute the gradient of the sigmoid function
def sigmoid_gradient(x):
    s = sigmoid(x)
    return s * (1 - s)

# Plot the gradient
x = np.linspace(-10, 10, 100)
gradient = sigmoid_gradient(x)

plt.plot(x, gradient)
plt.xlabel("x")
plt.ylabel("sigmoid'(x)")
plt.title("Gradient of Sigmoid Function")
plt.grid(True)
plt.show()
```

Now with the gradient we will comoute the heissan of x

```python
#Compute the Hessian of the sigmoid function
def sigmoid_hessian(x):
    s = sigmoid(x)
    return s * (1 - s) * (1 - 2 * s)

# Plot it
x = np.linspace(-10, 10, 100)
hessian = sigmoid_hessian(x)

plt.plot(x, hessian)
plt.xlabel("x")
plt.ylabel("sigmoid''(x)")
plt.title("Hessian of Sigmoid Function")
plt.grid(True)
plt.show()
```
