3.2.1 Limits and Continuity

Start with generating a function and plotting it for visualization

```python
# imports

import numpy as np
import matplotlib.pyplot as plt
import math as math#np.math is depreciated

# Generate x-axis
x = np.linspace(-20, 20, 400)

# Calculate corresponding y values using the f(x) function
y =np.sin(x)/x

# Create the plot
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of f(x)=sin(x)/x")
plt.grid(True)
plt.show()
```

```python
#Define f(x) as the functioon given above
def f(x):
  if x == 0:
    return 1
  else:
    return np.sin(x) / x

# Find the limit from both sides
limit_from_left = f(np.array([1e-10]))[0]
limit_from_right = f(np.array([-1e-10]))[0]
limit_from_left
limit_from_right
print("The limit is:", limit_from_left, "at x=0")
```

Thus we found the limit of the function above

3.2.2 Derivatives

We will come up with 2 sample functions, plot them, and then plot their derivatives.  This will help with visualization that the graph of the derivatives is the graph of slopes

```python
# Generate x-axis
x = np.linspace(-5, 5, 400)

# Calculate y = x^2
y_squared = x**2

# Create the first plot
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.plot(x, y_squared)
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.title("Plot of y = x^2")
plt.grid(True)


# Calculate y = sin(x)
y_sin = np.sin(x)

# Create the second plot
plt.subplot(2, 1, 2)
plt.plot(x, y_sin)
plt.xlabel("x")
plt.ylabel("y = sin(x)")
plt.title("Plot of y = sin(x)")
plt.grid(True)
plt.tight_layout()

# Generate the plots
plt.show()
```

Now we will plot the derivatives of the above two functions

```python
# Label the derivative of x^2
dy1 = 2 * x

# label the derivative of sin(x)
dy2 = np.cos(x)

# Create the first plot
plt.figure(figsize=(8, 6))
plt.subplot(2, 1, 1)
plt.plot(x, dy1)
plt.xlabel("x")
plt.ylabel("dy/dx = 2x")
plt.title("Derivative of y = x^2")
plt.grid(True)

# Create the second plot
plt.subplot(2, 1, 2)
plt.plot(x, dy2)
plt.xlabel("x")
plt.ylabel("dy/dx = cos(x)")
plt.title("Derivative of y = sin(x)")
plt.grid(True)
plt.tight_layout()

# Display the plots
plt.show()
```

Thus the derivatives of the two functions above have been plotted.

3.2.3 Taylors Theorem

The example of taylors theorem will be of the function from the first part of this assignment: Sin(x)/x.  The graph will have plotted 1st through fifth order approximation.

```python
#Define the function to solve for the taylor functions(assistance from online)
def taylor_approximation(x, n):
    approximation = 1.0
    for i in range(1, n + 1):
        term = (-1)**i * x**(2*i) / math.factorial(2*i + 1)
        approximation += term
    return approximation

# Calculate Taylor approximations for 1st through 5th order
y_exact = np.sin(x) / x
y_exact[x==0] = 1
y_1st = taylor_approximation(x, 1)
y_2nd = taylor_approximation(x, 2)
y_3rd = taylor_approximation(x, 3)
y_4th = taylor_approximation(x, 4)
y_5th = taylor_approximation(x, 5)

# Create the plot
x = np.linspace(-5, 5, 400)
plt.figure(figsize=(10, 6))
plt.plot(x, y_exact, label="Exact function")
plt.plot(x, y_1st, label="1st order")
plt.plot(x, y_2nd, label="2nd order")
plt.plot(x, y_3rd, label="3rd order")
plt.plot(x, y_4th, label="4th order")
plt.plot(x, y_5th, label="5th order")


plt.xlabel("x")
plt.ylabel("y")
plt.title("Taylor Approximations of sin(x)/x")
plt.grid(True)
plt.legend()
plt.show()
```

And above the 5 different order approximations of the function can be seen.
