2.3.1 Joint Probability Distributions

We will find the joint probability of 4 different objects with independent lifetimes.

The lifetimes will be of an exponential distribution

```python
#imports
import numpy as np
import pandas as pd

#set the parameters of each lifetime
param1=200
param2=400
param3=600
param4=800

#print the parameters
print("Parameter 1: ", param1)
print("Parameter 2: ", param2)
print("Parameter 3: ", param3)
print("Parameter 4: ", param4)
```

Based on the parameters for each item we will find the probability that the lifetimes of all the items is 550 days

```python
#set the expected lifetime
expected_lifetime=550

#calculate the probability using joint distribution
probability=np.exp(-param1/expected_lifetime)*np.exp(-param2/expected_lifetime)*np.exp(-param3/expected_lifetime)*np.exp(-param4/expected_lifetime)

#print the probability
print("The probability that the lifetimes of all the items is 550 days is: ", probability)
```

Thus joint probabiltiy distribution of 4 different random variables has been shown

2.3.2 Correlation and Dependence

We will generate three scatter plots with 1, 0, and -1 correlation

```python
# import
import matplotlib.pyplot as plt

# Generate data with correlation 1 with 50 points
np.random.seed(0)
x1 = np.random.rand(50)
y1 = x1

# Generate data with correlation 0 with 50 point
x2 = np.random.rand(50)
y2 = np.random.rand(50)

# Generate data with correlation -1 with 50 points
x3 = np.random.rand(50)
y3 = 1 - x3

```

With the data generated, the three plots will now be generated

```python
# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot 1
axes[0].scatter(x1, y1)
axes[0].set_title('Correlation = 1')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

# Plot 2
axes[1].scatter(x2, y2)
axes[1].set_title('Correlation = 0')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')

# Plot 3
axes[2].scatter(x3, y3)
axes[2].set_title('Correlation = -1')
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y')

plt.tight_layout()
plt.show()
```

And thus there are plots for 1, 0, and -1 correlation as shown above
