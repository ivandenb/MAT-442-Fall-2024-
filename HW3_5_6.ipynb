**3.5 and 3.6 K-means and Support vector machine**

We will start with an example of K-means

Wnat to take groups of data and put them into k different clusters

Start by making sample data

```python
# imports

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

#Sample data

X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

Plot the sample data

```python
#Plot the Data

plt.scatter(X[:,0], X[:,1], s=50)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Sample Data")
plt.show()
```

The gather the parameters for the k clusters

```python
# Initialize the 4 means of the clusters
kmeans = KMeans(n_clusters=4, random_state=0)

# Fit them to the means
kmeans.fit(X)

# Get labels at the data
labels = kmeans.labels_

# Get the center of the cluster
centroids = kmeans.cluster_centers_
```

Plot the K clusters

```python
# Plot the data points and clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='red', label='Centroids')
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

And thus we have an example of K-means
